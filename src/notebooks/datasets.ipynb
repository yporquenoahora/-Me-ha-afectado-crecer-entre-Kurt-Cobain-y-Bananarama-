```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZEVXbNG2KDcFcKOF?si=1333723a6eff4b7f"
playlist_URI ="6KOwiWg5zwrt83nEcx7HyI"
track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
playlist_len = len(track_uris)
```

```python
track_uris = np.empty((playlist_len), dtype=object)
track_ids = np.empty((playlist_len),  dtype=object)
track_names = np.empty((playlist_len),  dtype=object)

aAnalisysTrack = []
aAnalisysBeats = []
aAnalisysSections = []
aAnalisysSegments = []
aAnalisysTatums = []


playlist_tracks = sp.playlist_tracks(playlist_URI)["items"]

tracks = []
for i in playlist_tracks:
    t = i.get("track")
    album_release = t.get("album").get("release_date")    
    t.update({"release_date": album_release})
    tracks.append(t)
dfTracks = pd.DataFrame(tracks)
dfTracks.head()
```

```python
aFeatures = []
dfFeatures = {}
for ix,track in enumerate(playlist_tracks):
 
    #URI
    track_uri = track["track"]["uri"]
    track_uris[ix] = track["track"]["uri"]
    track_id = track_uri.split(":")[2]
    track_ids[ix] = track_id
    
    #Track name
    track_name = track["track"]["name"]
    track_names[ix] = track_name
    
    #Main Artist
    artist_uri = track["track"]["artists"][0]["uri"]
    artist_info = sp.artist(artist_uri)

    
    #Name, popularity, genre
    artist_name = track["track"]["artists"][0]["name"]
    artist_pop = artist_info["popularity"]
    artist_genres = artist_info["genres"]
    
    #Album
    album = track["track"]["album"]["name"]
    release_date = track["track"]["album"]["release_date"]
    
    #Popularity of the track
    track_pop = track["track"]["popularity"]
    
    #features.append(sp.audio_features(track_uri))
    features = sp.audio_features(track_uri)[0]
    features.update({"id":track_id, "names":track_name,"artists":artist_name, "genre": artist_genres, "release_date": release_date, "popularity": artist_info["popularity"]})
    
    
    #analisys = sp.audio_analysis(track_uri)
    

    aFeatures.append(features)
        

#df = pd.DataFrame({ "names": track_names, "artists":artist_name, "id":  track_id})
df = pd.DataFrame(aFeatures, columns = ["names", "artists", "release_date","id", "genre", "popularity"])

df.head()
```

```python
#canciones por genero general


sGenres = df["genre"]
genres_list = set()

for i,r in sGenres.items():       
        for item in r:          
          genres_list.add(item)
          
print(genres_list)

res = {}
for index, row in df.iterrows():       
    for gen in row.genre:
        res[gen] = (row.name)
            
dfGenres = pd.DataFrame([res])
dfGenres.to_csv("../data/songs_genres.csv")
dfGenres
```

```python
# Shows the top tracks for a user

import spotipy
from spotipy.oauth2 import SpotifyOAuth

scope = 'user-top-read'

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id = 'your client_id'
                                                client_secret = 'your client_secret'
                                               redirect_uri="http://localhost:1234",
                                               scope='user-top-read user-follow-read'))


ranges = ['short_term', 'medium_term', 'long_term']
subset = []
for sp_range in ranges:
    #print("range:", sp_range)
    results = sp.current_user_top_tracks(time_range=sp_range, limit=50)
    for i, item in enumerate(results['items']):                 
        artist_uri = item["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)        
        artist_genres = artist_info["genres"]
        subset.append([item["name"],item['artists'][0]['name'], item["album"]["release_date"],sp_range,item["id"], artist_genres,artist_info["popularity"]])

aFeatures = []
dfFeatures = {}
print(subset)

for ix,track in enumerate(subset):
    features = sp.audio_features(track[4])[0]
    aFeatures.append(features)

df = pd.DataFrame(aFeatures)

dfMostListened = pd.DataFrame(subset,  columns=["names", "artists","release_date", "range", "id", "genre", "popularity" ])

myConcatDF = pd.concat([dfMostListened, df], axis=1)
print(myConcatDF.columns)
myConcatDF.to_csv("../data/my_most_listened.csv")
myConcatDF.head()
```
